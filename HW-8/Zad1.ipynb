**Задача 1. Даны значения величины заработной платы заемщиков банка (zp) и значения их поведенческого кредитного скоринга (ks):\
zp = [35, 45, 190, 200, 40, 70, 54, 150, 120, 110]\
ks = [401, 574, 874, 919, 459, 739, 653, 902, 746, 832]\
Найдите ковариацию этих двух величин с помощью элементарных действий, а затем с помощью функции cov из numpy
Полученные значения должны быть равны.\
Найдите коэффициент корреляции Пирсона с помощью ковариации и среднеквадратичных отклонений двух признаков, а затем с использованием функций из библиотек numpy и pandas.**

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

zp = np.array([35, 45, 190, 200, 40, 70, 54, 150, 120, 110])
ks = np.array([401, 574, 874, 919, 459, 739, 653, 902, 746, 832])
```

**Решение:**\
Сначала найдем ковариацию двух величин:

```python
cov = np.mean(zp * ks) - np.mean(zp) * np.mean(ks)
print(f'{cov: .2f}')
print(np.cov(zp,ks,ddof=0))
```

Затем найдем коэффициент Пирсона с помощью ковариации по формуле:\
$r_{zp,ks} = \frac {cov_{zp,ks}}{\sigma_{zp} * \sigma_{ks}}$


помня, что значение ковариации мы считали по смещенному отклонению:

```python
stdX = np.std(zp, ddof=0)
stdY = np.std(ks, ddof=0)

corr = cov / (stdX * stdY)
print(f'Расчетный коэффициент Пирсона равен: {corr: .4f}')
```

```python
print(np.corrcoef(zp,ks))
```

Вывод: коэффициент корреляции ~0,89 показывает, что присутствует сильная линейная зависимость. В этом можно убедиться, построив график:

```python
plt.scatter(zp,ks)
plt.xlabel('Заработная плата')
plt.ylabel('Поведенческий кредитный скоринг')
plt.show
```
