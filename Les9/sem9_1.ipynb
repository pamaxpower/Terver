Постройте графики для приведенных ниже наборов данных. Найдите коэффициенты для линии регрессии и коэффициенты детерминации. Что вы замечаете? Нанесите на график модель линейной регрессии.

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

x1 = np.array([30,30,40,40])
y1 = np.array([37,47,50,60])

x2 = np.array([30,30,40,40,20,20,50,50])
y2 = np.array([37,47,50,60,25,35,62,72])

x3 = np.array([30,30,40,40,20,20,50,50,10,10,60,60])
y3 = np.array([37,47,50,60,25,35,62,72,13,23,74,84])
```

Построим график первого набора данных и посчитает коэффициенты по математическим формулам

```python
plt.scatter(x1,y1)

plt.show
b1 = (np.mean(x1*y1) - np.mean(x1) * np.mean(y1)) / (np.mean(x1**2) - np.mean(x1)**2)
b0 = np.mean(y1) - b1*np.mean(x1)
print(f'b1 = {b1}')
print(f'b0 = {b0}')
print(f'Готовая модель: y = {b0} + {b1} * x')

r = np.corrcoef(x1, y1)[1,0]
print(f'Коэффициент детерминации: {r**2: .4f}')

plt.plot(x1, b0 + b1*x1)
```

Построрим график второго набора данных и подсчитаем коэффициенты матричным методом:

```python
plt.scatter(x2,y2)
plt.show

xx2 = x2.reshape((8,1))
yy2 = y2.reshape((8,1))

X2 = np.hstack([np.ones((8,1)),xx2])
B2 = np.dot(np.linalg.inv(np.dot(X2.T, X2)), X2.T @ yy2)

print(f'b1 = {B2[1]}')
print(f'b0 = {B2[0]}')

r = np.corrcoef(x2, y2)[1,0]
print(f'Коэффициент детерминации: {r**2: .4f}')

plt.plot(x2, b0 + b1*x2)
```
