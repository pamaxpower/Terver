Проверить условия применимости линейной регрессии

```python
import matplotlib.pyplot as plt
import scipy.stats as stats
import numpy as np

x = np.array([10,8, 13, 9,11,14, 6,4,12, 7,5])
y = np.array([8.04, 6.95, 7.58, 8.81, 8.33, 9.96, 7.24, 4.26, 10.84, 4.82, 5.68 ])

y_hat = 3 + 0.5 * x
y_hat
```

```python
y_hat1 = y_hat.reshape(1,-1)
y_hat1
```

```python
res = y - y_hat1
res
```

```python
plt.scatter(x, res)
plt.show
```

```python
stats.shapiro(res)
```

p-value > $\alpha$, значит $H_0$  верна

```python
plt.scatter(y_hat, res)
```

Оценить статистическую значимость полученной модели.

```python
df1 = 2 - 1
df2 = 11 - 2
Ssf = sum((y_hat - np.mean(y))**2)
print(f'Ssf={Ssf}')
Sso = np.sum((y-y_hat)**2)
print(f'Ss0={Sso}')
Msf = Ssf / df1
print(f'Msf={Msf}')
Mso = Sso / df2
print(f'Mfo={Mso}')

F = Msf / Mso
print(f'Критерий Фишера: {F}')
```

```python
Ft = stats.f.ppf(1-0.05, df1, df2)
print(f'Критерий Фишера табличный: {Ft}')
```

Критическое значение 5.99, а расчетное 17,98, т.е. попадает в область принятия гипотезы $H_1$. Т.е. наша модель признана статистически значимой на уровне α= 0.05
