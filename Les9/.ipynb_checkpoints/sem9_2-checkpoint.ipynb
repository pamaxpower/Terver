Постройте графики по приведенным ниже данным. Для какого графика можно использовать модель линейной регрессии.

```python
import numpy as np
import matplotlib.pyplot as plt

x1 = np.array([10,8, 13, 9,11,14, 6,4,12, 7,5])
y1 = np.array([8.04, 6.95, 7.58, 8.81, 8.33, 9.96, 7.24, 4.26, 10.84, 4.82, 5.68 ])

y2 = np.array([ 9.14, 8.14, 8.74,8.77, 9.26, 8.10, 6.13, 3.10, 9.13, 7.26, 4.74])
y3 = np.array([7.46,6.77, 12.74, 7.11, 7.81, 8.84, 6.08, 5.39, 8.15, 6.42, 5.73])

x4 = np.array([8, 8, 8, 8, 8, 8, 8, 19, 8, 8, 8])
y4 = np.array([6.58, 5.76, 7.71, 8.84, 8.47, 7.04, 5.25,12.5, 5.56, 7.91, 6.89])

x0= np.array([ 10, 8, 13, 9, 11, 14, 6, 4, 12, 7,5, 15, 16, 18 ])
y0 = np.array([ 9.14, 8.14, 8.74,8.77, 9.26, 8.10, 6.13, 3.10, 9.13, 7.26, 4.74, 6.5, 5, 2.9])
```

Модель используется, если через данные можно визуально провести прямую

```python
plt.scatter(x1,y1)
plt.show
```

Первый набор данных подходит для использования модели линейной регрессии

```python
x = x1.reshape((11,1))
y = y1.reshape((11,1))

X = np.hstack([np.ones((11,1)),x])

B = np.dot(np.linalg.inv(np.dot(X.T, X)), X.T @ y)
print(f'b1 = {B[1]}')
print(f'b0 = {B[0]}')

r = np.corrcoef(x1, y1)[1,0]
print(f'Коэффициент детерминации: {r**2: .4f}')

plt.scatter(x1,y1)
plt.plot(x1, B[0] + B[1]*x1)
```

```python
plt.scatter(x1,y2)
plt.show
```

Второй набор данных очевидно не подходит для использования модели линейной регрессии. Тут на лицо нехватка данных (график может продолжить "путь параболы" и тогда это будет квадратичная зависимость)

```python
plt.scatter(x1,y3)
plt.show
```

Третий набор данных подходит для использования модели линейной регрессии. Здесь заметный очевидный выброс, а такие выбросы из анализа исключаются

```python
x = x1.reshape((11,1))
y = y3.reshape((11,1))

X = np.hstack([np.ones((11,1)),x])

B = np.dot(np.linalg.inv(np.dot(X.T, X)), X.T @ y)
print(f'b1 = {B[1]}')
print(f'b0 = {B[0]}')

r = np.corrcoef(x1, y3)[1,0]
print(f'Коэффициент детерминации: {r**2: .4f}')

plt.scatter(x1,y3)
plt.plot(x1, B[0] + B[1]*x1)
```

```python
plt.scatter(x4,y4)
plt.show
```

Четвертый набор данных не подходит для использования модели линейной регрессии. Тут есть подозрение на константу. А если это константа, то не понятно зачем вообще анализировать

```python
plt.scatter(x0,y0)
plt.show
```

Пятый набор данных не подходит. Очевидная квадратичная зависимость - нелинейная модель
